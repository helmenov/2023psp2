#　乱数を予測する

## 「乱数」とは

「次に出る1個の値が全く予測できない数値の系列」

つまり，癖が無い（偏りが無い）数値の系列．次に出る値を予測しても必ず失敗する数値の系列である．

## 癖がある数値の系列の例

癖がある数値系列の代表例として，漸化式で求まる級数がある．

..., 16, 32, 64, 128, 

という数値の系列は，連続して3個か4個くらい観察すれば，$a_{n+1}=2a_n$という等比級数だと予測がついて，次に出る値が予測できる．

128の次に出てくる値が実際には256ではなく，64かもしれない．そうすると64を観察した時点で予想を立て直して，次の値は32なんじゃないかと思うのではないだろうか．
少なくとも偶数なんじゃないかとか．

では，次の系列はどうだろう．

..., 16, 33, 68, 139, 

```
a_0, a_1, a_2, a_3 = 16, 33, 68, 139,
```

とする．$a_1 - a_0 = 17$, $a_2 - a_1 = 35$, $a_3 - a_2 = 71$である．この観察から，次に出る値は$a_4 =　282$と予測できるのではなかろうか．
